```python
# Next, redo the sorting on Gross

# Select all movies released after (>=) 2010 and with IMDB_Rating>=8.5
# Show their title, released year, Certificate, and gross.
# Sort them in descending order of "Gross"
(
    imdb[['Series_Title', 'Released_Year', 'Certificate', 'Gross']]
     [imdb['Released_Year'] >= 2010][imdb['IMDB_Rating'] >= 8.5]
    .sort_values('Gross', ascending=False)
)
```

```python
# Add a new column "Runtime_min" by removing the substring " min" in "Runtime"
# Set its data type as int
# Hint: https://stackoverflow.com/questions/36505847/substring-of-an-entire-column-in-pandas-dataframe
Runtime_min = imdb['Runtime'].str.replace(' min', '').astype(int)
imdb['Runtime_min'] = Runtime_min
imdb.head()
```

```python
# Add a new column "Age_Year" by expression: Current year - Released_Year
Age_Year = 2026 - imdb['Released_Year']
imdb['Age_Year'] = Age_Year
imdb.head()
```

```python
# Add a new column "Decade" with values as 1980, 1990, 2000, 2010, 2020, etc.
Decade = (imdb['Released_Year'] // 10) * 10
imdb['Decade'] = Decade
imdb.head()
```

## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies
imdb['Gross'].agg('sum')
```

```python
# Average "No_of_Votes" of all movies
imdb['No_of_Votes'].agg('mean')
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order
imdb.groupby(['Decade']).count()['Series_Title'].sort_values(ascending=False)
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
imdb['Director'].value_counts().head(10)
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
imdb.groupby(['Director']).agg({'Series_Title': 'count', 'IMDB_Rating': 'mean', 'Gross': 'sum'}).sort_values('Gross', ascending=False).head(10)
#
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count
imdb.groupby(['Decade', 'Director']).agg({'Series_Title': 'count', 'IMDB_Rating': 'mean'}).sort_values('Series_Title', ascending = False)
```

```python
# Find the top 3 highest rated movies of each year since 2010.
# Show their released year, ranking, title, and IMDB_Rating.
# Sort them in descending order of year and ascending order of ranking
imdb[['Released_Year', 'IMDB_Rating', 'Series_Title']][imdb['Released_Year'] >= 2010].sort_values(['Released_Year', 'IMDB_Rating'], ascending=[False, True]).groupby('Released_Year').head(3)
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.
imdb.groupby(['Director', 'Decade']).agg({'Series_Title': 'count'}).sort_values('Series_Title', ascending=False)
```

```python
# Ask two questions about this dataset and write your code to answer them.
# Question 1: Average IMDB Rating for each decade.
imdb.groupby(['Decade']).agg({'IMDB_Rating': 'mean'})
```

```python
# Questions 2: Top five highest grossing directors on the list.
imdb.groupby(['Director']).agg({'Gross': 'sum'}).sort_values('Gross', ascending=False).head(5)
```
